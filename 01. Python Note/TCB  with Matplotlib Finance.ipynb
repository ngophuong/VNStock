```python
import mplfinance as mpf
import pandas as pd
import numpy as np
import matplotlib as mpl
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/ngophuong/VNStock/main/excel_tcb.csv'
df = pd.read_csv(url, index_col = 0)
df['datetime']=pd.to_datetime(df['<DTYYYYMMDD>'],format='%Y%m%d')
df.set_index('datetime',inplace = True)
df = df.sort_index(ascending = True)
df.head(3)
```

```python
df =df[['<OpenFixed>','<HighFixed>','<LowFixed>','<CloseFixed>','<Volume>']]
df = df.tail(200)
```

```python
df = df.rename(columns={"<OpenFixed>": "Open", "<HighFixed>": "High",
                        "<LowFixed>":"Low","<CloseFixed>":"Close","<Volume>":"Volume"})
list(df.columns)
```

```python
mpf.plot(df.tail(0),type='candle',mav=(3,6,9),volume=True,show_nontrading=True)
```
